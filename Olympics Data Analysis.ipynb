```python
## Gold medals
gold_medals = athletes_df[(athletes_df.Medal == 'Gold')]
gold_medals.head()
```

```python
## take only values that are diffrent from NaN
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
```

```python
gold_medals['ID'][gold_medals['Age'] > 60 ].count()
```

```python
sport_event = gold_medals['Sport'][gold_medals['Age'] > 60]
sport_event
```

```python
DT_event = gold_medals[['Sport','Name','Sex','Team']][gold_medals['Age'] > 60]
DT_event
```

```python
## plot for sporting event
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sport_event)
plt.title("Gold medal for athletes over 60 per sport")
```

```python
## Gold medals for each country
gold_medals.Region.value_counts().reset_index(name='Medals').head()
```

```python
totalGoldMedals = gold_medals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x="index", y ="Medal", data=totalGoldMedals, height=5, kind ='bar', palette="rocket")
g.despine(left=True)
g.set_xlabels("top 6 countries")
g.set_ylabels("number of Medals")
plt.title('Gold Medals per country')
```
